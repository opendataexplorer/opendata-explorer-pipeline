```python
import numpy as np
import pandas as pd
import requests
import zipfile
from os import listdir, remove
from os.path import getsize
from pathlib import Path
```

# Download All Tables to data/input

```python
data_table = pd.read_json("https://www150.statcan.gc.ca/t1/wds/rest/getAllCubesList")
data_table
```

```python
for pid in data_table.productId:
    try:
        url = f"https://www150.statcan.gc.ca/n1/en/tbl/csv/{pid}-eng.zip"

        r = requests.get(url)

        with open(f'../data/input/{pid}-eng.zip', 'wb') as f:
            f.write(r.content)

        with zipfile.ZipFile(f'../data/input/{pid}-eng.zip', 'r') as zip_ref:
            zip_ref.extractall('../data/input')

        remove(f'../data/input/{pid}-eng.zip')
        remove(f'../data/input/{pid}_MetaData.csv')
        
    except Exception as e:
        print(e)
```

# Summary

```python
input_path = 'data/input'
files = listdir(input_path)
sizes = []
files_under_5mb = []
files_under_5mb_size = []
for file in files:
    sizes.append(getsize(f"{input_path}/{file}"))
    if getsize(f"{input_path}/{file}") < 5000000:
        files_under_5mb.append(file)
        files_under_5mb_size.append(getsize(f"{input_path}/{file}"))
```

```python
print(f"Total Size: {sum(sizes)/1000000000} GB")
print(f"Count: {len(files_under_5mb)}")
print(f"Files under 5mb Total Size: {sum(files_under_5mb_size)/1000000000} GB")
```
