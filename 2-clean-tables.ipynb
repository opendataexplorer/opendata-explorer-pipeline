```python
import pandas as pd
from os import listdir
from os.path import splitext
```

```python
input_path = "data/input"
output_path = "data/output"
```

```python
input_files = listdir(input_path)
```

```python
def clean_dataframes(file, resume_clean=0):
    pid = splitext(file)[0] 
    extension = splitext(file)[1]
    def clean_file(file, pid):
        
        input_csv = pd.read_csv(f"data/input/{file}", low_memory=False)
        cleaned_csv = input_csv.drop(columns=["DGUID", 
                                          "UOM_ID", 
                                          "SCALAR_ID", 
                                          "VECTOR", 
                                          "COORDINATE", 
                                          "STATUS", 
                                          "SYMBOL", 
                                          "TERMINATED",
                                        "DECIMALS"], 
                                      errors='ignore')
        # if cleaned_csv['REF_DATE'].astype(str).map(len).max() == 4:
        #     cleaned_csv['REF_DATE'] = cleaned_csv['REF_DATE'].astype(str) + "-01-01"
        # if cleaned_csv['REF_DATE'].astype(str).map(len).max() == 7:
        #     cleaned_csv['REF_DATE'] = cleaned_csv['REF_DATE'].astype(str) + "-01"
        cleaned_csv = cleaned_csv.rename(str.lower, axis='columns')
        cleaned_csv.columns = cleaned_csv.columns.str.replace(" ", "_")
        cleaned_csv.to_json(f'data/output/{pid}.json', orient='records')
    if extension == '.csv':
        if resume_clean == 0:
            try:
                clean_file(file, pid)        
            except Exception as e:
                print(e)
        elif resume_clean > 0:
            if int(pid) > resume_clean:
                try:
                    clean_file(file, pid)
                except Exception as e:
                    print(e)
```

```python
for file in input_files:
    clean_dataframes(file)
```
