```python
import pandas as pd
from tqdm import tqdm
from os import listdir
```

```python
output_path = 'data/output'
```

```python
files = listdir(output_path)
```

```python
check_columns = ['geo','ref_date','value']

for file in tqdm(files):
    check_csv = pd.read_json(f"data/output/{file}")
    
    # Dataframe is not empty
    if (check_csv.shape[0] == 0):
        print(f"{file} is empty")
    
    # Dataframe has columns: Geography, Reference Date, Value
    for column in check_columns:
        if column not in check_csv.columns.to_list():
            print(f"{file} does not contain column: {column}")

    # Ref Date is exactly 10 strings long when no slash
    if check_csv.ref_date.map(len).max()!=10 and not check_csv.ref_date.str.contains("\/").any():
        print(f"{file} has incorrect date")

    # Ref Date is exactly 9 strings long when slash
    if check_csv.ref_date.map(len).max()!=9 and check_csv.ref_date.str.contains("\/").any():
        print(f"{file} has incorrect date")
```
